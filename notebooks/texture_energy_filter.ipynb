```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

### Set up the repo paths

```python
repo_path = Path("/home/nico/workspace/github.com/Nico-Sander/KI-Project-WS2526")
print(f"Repository root: {repo_path}")
data_path = repo_path / "data" / "raw"
print(f"Data root: {data_path}")
```

### Load an image

```python
img_raw = cv2.imread(data_path /  "IMG_8449.jpeg")
img_gray = cv2.cvtColor(img_raw, cv2.COLOR_BGR2GRAY)

# Show the image
plt.imshow(img_gray, cmap='gray')
plt.show()
```

### Edge detection (sobel)

```python
grad_x = cv2.Sobel(img_gray, cv2.CV_32F, 1, 0, ksize=3)
grad_y = cv2.Sobel(img_gray, cv2.CV_32F, 0, 1, ksize=3)
mag = cv2.magnitude(grad_x, grad_y)
mag = cv2.convertScaleAbs(mag)

plt.imshow(mag, cmap='gray')
plt.show()
```

### blur

```python
blurred = cv2.blur(mag, (10, 10))

plt.imshow(blurred, cmap='gray')
plt.show()
```

### Second edge detection

```python
grad_x = cv2.Sobel(blurred, cv2.CV_32F, 1, 0, ksize=11)
grad_y = cv2.Sobel(blurred, cv2.CV_32F, 0, 1, ksize=11)
mag2 = cv2.magnitude(grad_x, grad_y)
mag2 = cv2.convertScaleAbs(mag)

plt.imshow(mag2, cmap='gray')
plt.show()
```
